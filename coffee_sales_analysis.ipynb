```python
def time_bucket(hour):
    if 6 < hour < 11:
        return "morning"
    if 12 < hour < 16:
        return "after noon"
    if 17 < hour < 21:
        return "evening"
    else :
        return "late hour"
df["time_bucket"] = df["hour"].apply(time_bucket)
```

```python
df.groupby("time_bucket")["revenue"].sum().sort_values(ascending=False)
```
